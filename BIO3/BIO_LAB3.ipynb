```python
edgeNumber = int(input("Enter number of edges: "))
#making list of all edges
edges = []
for i in range(edgeNumber):
    edge = input("Enter edge as 'A B': ")
    x, y = edge.split(" ")[0], edge.split(" ")[1]
    l1 = [x, y]
    edges.append(l1)

#making list of all nodes
allNodes = []
for edge in edges:
    #['a', 'b']
    if(edge[0] not in allNodes):
        allNodes.append(edge[0])
    elif(edge[1] not in allNodes):
        allNodes.append(edge[1])

#initializing graphList with nodes
graphList = {}
for node in allNodes:
    graphList[node] = []

#making the graphList
for edge in edges:
    l2 = []
    if(not len(graphList.get(edge[0])) == 0):
        e = graphList.get(edge[0])
        for element in e:
            l2.append(element)
    l2.append(edge[1])
    graphList[edge[0]] = l2

#printing
print("The edges are ")
for vertices in graphList:
    allEdges = graphList.get(vertices)
    for edge in allEdges:
        print(vertices + " → " + edge, end = "\t")
    if(not allEdges == []):
        print()
```

```python
file = open("data.txt", 'r')
allEdges = []
for data in file:
    edge = data.split(" ")
    edge[2] = edge[2].replace("\n", "")
    allEdges.append(edge)
    allEdges.append([edge[1], edge[0], edge[2]])
print(allEdges)
vertices = []
for edge in allEdges:
    if edge[0] not in vertices:
        vertices.append(edge[0])
    if edge[1] not in vertices:
        vertices.append(edge[1])

graphList = {}
for vertex in vertices:
    graphList[vertex] = []

for edge in allEdges:
    l1 = []
    if not len(graphList.get(edge[0])) == 0:
        e = graphList.get(edge[0])
        for element in e:
            l1.append(element)
    x = [edge[1], edge[2]]
    l1.append(x)
    graphList[edge[0]] = l1

print("The vertices are: ", *graphList.keys())
print("Edges are: ")
for vertex in graphList:
    edges = graphList.get(vertex)
    for edge in edges:
        print(vertex, "<-->", edge[0], "=> weight:", edge[1], end = ";\t")
    print()
```
