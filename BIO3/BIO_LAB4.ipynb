```python
#Create a list of all kmers
seq = "GTACGTACGAT"
overlap = 4
kmers = [seq[i:i+k] for i in range(len(seq)-k+1)]
print(kmers)
```

```python
#represent each k-mer as node
import networkx as nx
G = nx.Graph()
for kmer in kmers:
    G.add_node(kmer)
```

```python
#connect two nodes with an edge if they overlap
for i in range((len(kmers))):
    for j in range((len(kmers))):
        if kmers[i][overlap:] == kmers[j][:overlap]:
            G.add_edge(kmers[i], kmers[j], weight=overlap)
```

```python
#visualize the graph
import matplotlib.pyplot as plt
nx.draw(G, with_labels=True)
plt.show()
```
