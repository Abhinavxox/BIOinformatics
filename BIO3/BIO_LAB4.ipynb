```python
#Create a list of all kmers
seq = "GTACGTACGAT"
overlap = 4
k=6
kmers = [seq[i:i+k] for i in range(len(seq)-k+1)]
print(kmers)
```

```python
#represent each k-mer as node
import networkx as nx
G = nx.Graph()
for kmer in kmers:
    G.add_node(kmer)
```

```python
#visualize the graph
import matplotlib.pyplot as plt
for k in kmers:
  for j in kmers:
    if k[-5:] == j[0:5]:
      G.add_edge(k,j,)
      continue
    if k[-4:]==j[0:4]:
      G.add_edge(k,j,weight=4)
plt.figure(figsize=(20,10))
pos=nx.circular_layout(G)
nx.draw_networkx(G,pos=pos,with_labels=True)
nx.draw_networkx_edge_labels(G,pos)
```
