```python
#problem1
#Find the reverse complement of a DNA string
dna_sequence = input("Enter a DNA sequence").upper()

dict = {"T":"A",
        "G":"C",
        "A":"T",
        "C":"G"}
compliment=""
for x in dna_sequence:
  compliment = compliment + dict[x]

print(compliment[::-1])

```

```python
#problem2
#compute the number of times a pattern appears in a text
dna = input("Enter a DNA sequence ").upper()
pattern =  input("Pattern ").upper()
length = len(pattern)
flag = False
pointer = 0
for x in range(len(dna)):
  if pattern[0]==dna[x]:
    if pattern == dna[x:x+length]:
      flag = True
    else: 
      continue
  if flag:
    pointer = pointer+1
    flag= False
print(pointer)
```

```python
#problem3
#find all occurrences of a pattern in a dna string
dna = input("Enter a DNA sequence ").upper()
pattern =  input("Pattern ").upper()
length = len(pattern)
flag = False
z = ""
for x in range(len(dna)):
  if pattern[0]==dna[x]:
    if pattern == dna[x:x+length]:
      flag = True
    else: 
      continue
  if flag:
    z = z + str(x)
    z = z + " "
  flag= False
print(z)
      
```

```python
#problem4
#find the most frequend words in a string
dna = input("Enter a DNA sequence ").upper()

k_mer = int(input("Enter k"))

list = []
length = len(dna)
for i in range(length):
  if(i == length-3):
    break
  list.append(dna[i:i+k_mer])

#now getting count of each element in the list
dict={}
for i in list:
  if i in dict.keys():
    temp = dict[i]
    dict[i] = temp+1
  else:
    dict.update({i:0})
print(dict)

#getting the highest count in dict
pointer = 0
for i in dict.values():
  if i>pointer:
    pointer= i

output = ""
for key in dict.keys():
  if dict[key] == pointer:
    output = output + key
    output = output + " "

print(output)
 
```

```python
dna = input("Enter a dna")
k = int(input("Enter k"))
dict = {}
for x in range(0,len(dna)-k):
  count = 0
  pattern = dna[x:x+k]
  for y in range(0,len(dna)-k):
    if dna[y:y+k]==pattern:
      count=count+1
  dict[pattern]=count
print(dict)

output = ""
counter = 0
for keys in dict.keys():
  if dict[keys]>counter:
    output = keys
    counter = dict[keys]

print(output)
```
