```python
#question1
#generate a random dna sequence
import random, time

l = random.randint(1, 1000)
list = ['A', 'C', 'G', 'T']
dna =  ""
while(l!=0):
    dna += random.choice(list)
    l=l-1
print(dna)


start_time = time.time()
print("A : " + str(dna.count("A")))
print("C : " + str(dna.count("C")))
print("G : " + str(dna.count("G")))
print("T : " + str(dna.count("T")))
end_time = time.time()
print("The total time taken is :" , end=" ") 
print(end_time -  start_time)
```

```python
#question2
#compute the hamming distance between two strings

first = input("Input first dna string").upper()
second = input("Input second dna string").upper()

for i in range(0,  len(first)):
  distance = 0
  if(first[i] != second[i]):
      distance += 1

print("The required hamming distance is "+str(distance))
```

```python
#problem3
#find patterns forming clumps in a string
from collections import defaultdict

def search(inseq, k, L, t):
    lookup = defaultdict(list)
    result = set()
    
    for cursor in range(len(inseq) - k + 1):
        seg = inseq[cursor:cursor + k]
        
        # remove prior positions of the same segment
        # if they are more than L distance far
        while lookup[seg] and cursor + k - lookup[seg][0] > L:
            lookup[seg].pop(0)
        
        lookup[seg].append(cursor)
        if len(lookup[seg]) == t:
            result.add(seg)
    
    return result

DNA = input("What is the genomic sequence? ").upper()
k = int(input("input k"))
L = int(input("length of clump l"))
t = int(input("input t"))

result = search(DNA, k, L, t)
s = ""
for i in result:
  s = s+i+" "
print(s)
```

```python
#question4
#find a position in a genome minimising the skew
dna = input("enter dna")
c = 0
g = 0
min_skew = 0
skew_list = []
index = 0
for i in dna:
  index += 1
  if i == 'C':
    c += 1
  if i == 'G':
    g += 1
  skew = g-c
  if skew < min_skew:
    skew_list = [index]
    min_skew = skew
  if skew == min_skew and index not in skew_list:
    skew_list.append(index)
print(skew_list)
```
